```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Завантажте дані про рак грудей
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
X_standardized = StandardScaler().fit_transform(X)

# Визначте PCA з двома головними компонентами
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_standardized)
```

```python
components_df = pd.DataFrame(data=principal_components, columns=['First principal component', 'second principal component'])
components_df['Target'] = y
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(components_df['First principal component'], components_df['second principal component'], c=components_df['Target'], cmap='viridis')
plt.title('Breast Cancer Dataset - PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
